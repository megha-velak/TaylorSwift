```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("taylor_swift_spotify.csv",index_col=0)
df.head()
```

```python
df.describe()
```

```python
# Distribution of numeric features
numeric_features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 
                    'liveness', 'loudness', 'speechiness', 'tempo', 'valence', 
                    'popularity', 'duration_ms']
df[numeric_features].hist(bins=20, figsize=(15, 10))
plt.suptitle("Distribution of Numeric Features", y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df[numeric_features].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
top_ten_songs = df.sort_values(by='popularity', ascending=False).head(10)

# Plot top ten most popular songs
plt.figure(figsize=(10, 6))
plt.barh(top_ten_songs['name'], top_ten_songs['popularity'],color="purple")
plt.xlabel('Popularity')
plt.ylabel('Song')
plt.title('Top Ten Most Popular Songs')
plt.gca().invert_yaxis()  # Invert y-axis to display the most popular song at the top
plt.tight_layout()
plt.show()
```
